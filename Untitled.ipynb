```python
from datasets import load_dataset
import pandas as pd
import lizard

dataset = load_dataset("THUDM/humaneval-x", "python", split="test")

# Convert to DataFrame
df = pd.DataFrame(dataset)
# Initialize lists to store metrics
input_code = []
line_counts = []
char_lengths = []
cc_values = []

# Compute metrics
for entry in dataset:
    # Combine declaration + canonical solution
    code = entry['declaration'] + entry['canonical_solution']
    input_code.append(code)
    # Line count
    line_counts.append(len(code.splitlines()))

    # Character length
    char_lengths.append(len(code))

    # Cyclomatic complexity
    analysis = lizard.analyze_file.analyze_source_code("inline.py", code)
    if analysis.function_list:
        cc = analysis.function_list[0].cyclomatic_complexity
    else:
        cc = None
    cc_values.append(cc)

# Add metrics to DataFrame
df['input_code'] = input_code
df['line_count'] = line_counts
df['code_length'] = char_lengths
df['cyclomatic_complexity'] = cc_values

df = df[
    (df['cyclomatic_complexity'] >= 4) &
    (df['cyclomatic_complexity'] <= 8) &
    (df['line_count'] < 15)
]

df.to_csv("filtered_humaneval.csv", index=False)
```
